# Modul 3: Struktur Kontrol Percabangan

- Sub-CPMK 2.1: Menyusun program dengan `if`/`elif`/`else` untuk percabangan

## Sarana dan prasarana

Python 3.10+, editor, terminal.

## Prosedur kerja

- Pelajari pola percabangan dan `match` pengenalan.
- Kerjakan soal bertahap dari sederhana ke bersarang.
- Verifikasi dengan beberapa kasus uji manual.

## Sintaks Percabangan if

Struktur dasar percabangan di Python:

```python
nilai = int(input("Nilai: "))
if nilai >= 80:
    print("A")
elif nilai >= 70:
    print("B")
elif nilai >= 60:
    print("C")
else:
    print("E")
```

Indentasi menandai blok yang termasuk dalam cabang `if`. Tidak ada tanda kurung kurawal seperti bahasa lain --- indentasi **adalah** struktur program.

## Kondisi Bertingkat dan Bersarang

Percabangan dapat bersarang untuk logika yang lebih kompleks. Hindari terlalu dalam tanpa komentar; pertimbangkan memecah ke fungsi pada modul berikutnya.

```python
x = int(input("x: "))
if x >= 0:
    if x == 0:
        print("nol")
    else:
        print("positif")
else:
    print("negatif")
```

## Ekspresi Kondisional (Ternary)

Python mendukung bentuk singkat `a if kondisi else b` yang menghasilkan nilai berdasarkan kondisi:

```python
x = int(input("x: "))
label = "genap" if x % 2 == 0 else "ganjil"
print(x, "adalah", label)
```

Prioritas logika harus jelas: gunakan tanda kurung pada kondisi kompleks agar mudah dibaca.

## Pernyataan pass

Pernyataan `pass` tidak melakukan apa-apa. Berguna sebagai *placeholder* ketika sintaks mensyaratkan sebuah pernyataan tetapi program belum memerlukan aksi apapun. Ini sering digunakan selama pembuatan kode (prototyping).

```python
def fungsi_belum_jadi():
    pass   # akan diimplementasikan nanti

# Dalam blok if yang belum diisi:
pilihan = input("Pilih (a/b): ")
if pilihan == "a":
    print("Anda memilih a")
elif pilihan == "b":
    pass   # TODO: implementasikan opsi b
else:
    print("Pilihan tidak valid")
```

## Pernyataan match

Sejak Python 3.10, tersedia pernyataan `match` sebagai alternatif dari rantai `if`/`elif` yang panjang. `match` mencocokkan nilai dengan sejumlah *pattern* (pola) dan membuat kode lebih mudah dibaca.

```python
perintah = input("Perintah (berhenti/mulai/jeda): ")
match perintah:
    case "berhenti":
        print("Program dihentikan.")
    case "mulai":
        print("Program dimulai.")
    case "jeda":
        print("Program dijeda.")
    case _:              # wildcard, cocok dengan apapun
        print(f"Perintah '{perintah}' tidak dikenal.")
```

### match dengan Beberapa Pola dalam Satu Case

Beberapa pola dapat digabung menggunakan `|` (OR):

```python
kode_status = 404
match kode_status:
    case 200 | 201 | 204:
        print("Sukses")
    case 400:
        print("Permintaan tidak valid")
    case 404:
        print("Tidak ditemukan")
    case 500 | 503:
        print("Kesalahan server")
    case _:
        print("Kode tidak dikenal")
```

### Kode: akar_persamaan_kuadrat.py

```python
"""Mencari akar persamaan kuadrat (rumus benar; kode Pascal sumber berisi typo)."""

import math


def main() -> None:
    print("    Mencari Akar Persamaan Kuadrat     ")
    print("      Persamaan Umum : ax^2+bx+c       ")
    print()
    a = float(input("a = "))
    b = float(input("b = "))
    c = float(input("c = "))
    print()
    d = (b * b) - (4 * a * c)
    print("Nilai Diskriminan :", d)
    print("Jenis Akar : ", end="")
    if d == 0:
        x1 = (-b) / (2 * a)
        print("Kembar")
        print(f"x1=x2={x1:10.2f}")
    elif d > 0:
        # Pascal salah menulis kedua akar dengan +sqrt(D); di sini dipakai + dan -
        x1 = (-b + math.sqrt(d)) / (2 * a)
        x2 = (-b - math.sqrt(d)) / (2 * a)
        print("Berlainan")
        print(f"x1 = {x1:10.2f}")
        print(f"x2 = {x2:10.2f}")
    else:
        # sqrt(-D), bukan -d kecil (typo Pascal)
        x1 = (-b) / (2 * a)
        x2 = math.sqrt(-d) / (2 * a)
        print("2 akar imajiner berlainan")
        print(f"x1 = {x1:5.2f}+{x2:5.2f}i")
        print(f"x2 = {x1:5.2f}-{x2:5.2f}i")


if __name__ == "__main__":
    main()
```

### Kode: huruf_mutu.py

```python
"""Konversi nilai akhir ke huruf mutu A-E."""


def main() -> None:
    quiz = 40.0
    absen = 100.0
    uts = 60.0
    uas = 50.0
    tugas = 80.0
    print(f"Absen = {absen:5.2f} UTS = {uts:5.2f}")
    print(f"Tugas = {tugas:5.2f} UAS = {uas:5.2f}")
    print(f"Quiz = {quiz:5.2f}")
    nilai = (
        (0.1 * absen) + (0.2 * tugas) + (0.3 * quiz) + (0.4 * uts) + (0.5 * uas)
    ) / 2
    huruf_mutu = ""
    if 85 < nilai <= 100:
        huruf_mutu = "A"
    elif 70 < nilai <= 85:
        huruf_mutu = "B"
    elif 55 < nilai <= 70:
        huruf_mutu = "C"
    elif 40 < nilai <= 55:
        huruf_mutu = "D"
    elif 0 <= nilai <= 40:
        huruf_mutu = "E"
    print("Huruf Mutu : ", huruf_mutu)


if __name__ == "__main__":
    main()
```

# Modul 4: Struktur Kontrol Perulangan

- Sub-CPMK 2.2: Menyusun program dengan perulangan `for` dan `while` termasuk pola bercabang

## Sarana dan prasarana

Python 3.10+, editor, terminal.

## Prosedur kerja

- Praktik `for`, `while`, `break`/`continue` dari contoh.
- Kerjakan latihan loop bersarang sesuai aktivitas.
- Dokumentasikan pola iterasi pada laporan singkat.

## Perulangan for

```python
# Iterasi pada range
for i in range(5):
    print(i, end=" ")
# 0 1 2 3 4

# Iterasi langsung pada string
for huruf in "Python":
    print(huruf, end="-")
# P-y-t-h-o-n-

# Iterasi pada list
buah = ["apel", "jeruk", "mangga"]
for b in buah:
    print(b)
```

Saat melakukan iterasi dan perlu memodifikasi koleksi, lebih aman untuk bekerja pada *salinan*: `for b in buah[:]:`. Tutorial Python resmi (§4.2) juga menyarankan menggunakan fungsi seperti `enumerate()` untuk mendapatkan indeks sekaligus nilai (dibahas di modul 6).

## Fungsi range()

```python
list(range(5))          # [0, 1, 2, 3, 4]
list(range(2, 8))       # [2, 3, 4, 5, 6, 7]
list(range(0, 20, 5))   # [0, 5, 10, 15]
list(range(10, 0, -2))  # [10, 8, 6, 4, 2]  -- menurun
```

## Perulangan while

Cocok bila jumlah iterasi tidak pasti:

```python
n = int(input("n: "))
i = 1
while i <= n:
    print(i * i)
    i += 1
```

## break

```python
# continue: cetak bilangan ganjil saja
for x in range(10):
    if x % 2 == 0:
        continue
    print(x, end=" ")
# 1 3 5 7 9

# break: berhenti saat menemukan bilangan negatif
angka = [3, 7, -1, 5, 2]
for a in angka:
    if a < 0:
        print(f"\nDitemukan angka negatif: {a}, berhenti.")
        break
    print(a, end=" ")
```

## Klausa else

Ini adalah fitur **unik** Python yang sering mengejutkan programmer dari bahasa lain. Blok `else` pada loop dieksekusi **setelah loop selesai secara normal** (tidak dihentikan oleh `break`). Sangat berguna untuk algoritme pencarian.

```python
# Mencari bilangan prima (contoh dari tutorial resmi)
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, "=", x, "*", n//x)
            break
    else:
        # Loop tidak menemukan faktor -> n adalah prima
        print(n, "adalah bilangan prima")
```

```python
daftar = [5, 3, 8, 1, 9, 2]
target = 7
i = 0
while i < len(daftar):
    if daftar[i] == target:
        print(f"Ditemukan {target} di indeks {i}")
        break
    i += 1
else:
    # Keluar dari while tanpa break -> target tidak ditemukan
    print(f"{target} tidak ada dalam daftar")
```

Pola `for/while ... else` adalah idiom Python yang elegan untuk pencarian: jika loop selesai tanpa `break`, berarti kondisi pencarian tidak terpenuhi dan blok `else` dieksekusi.

## pass

Seperti pada percabangan, `pass` dapat digunakan dalam loop sebagai placeholder:

```python
# Loop yang belum diimplementasikan isinya
while True:
    pass  # Placeholder - akan diisi dengan logika nanti
```

## Loop Bersarang

Digunakan untuk pola matriks sederhana, pola bintang, atau kombinatorika kecil:

```python
n = 5
for baris in range(1, n + 1):
    print("*" * baris)
# *
# **
# ***
# ****
# *****
```

### Kode: tabel_pangkat_x.py

```python
"""Tabel x, 1/x, x^2, x^3 untuk x = 1..10."""


def main() -> None:
    print("  x      1/x        x^2      x^3")
    print("-----------------------------------")
    x = 1
    while x <= 10:
        print(f"{x:5d}", end="")
        print(f"{1.0 / x:8.5f}", end="")
        pangkat = 2
        while pangkat <= 3:
            hasilpangkat = 1
            for _i in range(1, pangkat + 1):
                hasilpangkat *= x
            print(f"{hasilpangkat:8d}", end="")
            pangkat += 1
        print()
        x += 1


if __name__ == "__main__":
    main()
```

# Modul 5: Fungsi dan Modularitas

- Sub-CPMK 3.1: Mendefinisikan fungsi dengan parameter, nilai balik, dan ruang lingkup variabel lokal
- Sub-CPMK 3.2: Menggunakan argumen variadik (`*args`, `**kwargs`), ekspresi *lambda*, dan anotasi tipe dasar pada fungsi

## Sarana dan prasarana

Python 3.10+, editor, terminal.

## Prosedur kerja

- Definisikan fungsi dengan parameter dan nilai balik; uji dari REPL.
- Tambahkan `*args`/`**kwargs`, *lambda*, dan anotasi tipe sesuai contoh modul.
- Refaktor program monolitik menjadi fungsi-fungsi kecil.

## Definisi Fungsi

Fungsi memecah program menjadi bagian yang dapat digunakan kembali. Didefinisikan dengan `def`:

```python
def kuadrat(x):
    """Mengembalikan kuadrat dari x."""
    return x * x

def salam(nama):
    """Mencetak salam kepada nama."""
    print("Halo,", nama)

print(kuadrat(5))   # 25
salam("Aulia")      # Halo, Aulia
```

Pernyataan **pertama** dari badan fungsi bisa berupa *docstring* (dokumentasi fungsi dalam string literal). Gunakan `help(kuadrat)` di REPL untuk membacanya.

## Parameter dan Nilai Balik

Fungsi dapat mengembalikan beberapa nilai sekaligus dalam bentuk tuple:

```python
def statistik(data):
    """Mengembalikan (minimum, maksimum, rata-rata)."""
    return min(data), max(data), sum(data) / len(data)

nilai = [75, 90, 60, 85, 78]
mn, mx, rata = statistik(nilai)
print(f"Min={mn}, Max={mx}, Rata={rata:.2f}")
```

## Nilai Argumen Default

Parameter dapat memiliki nilai *default* yang digunakan bila argumen tidak diberikan saat pemanggilan (Tutorial Python §4.9.1):

```python
def hitung_pangkat(basis, eksponen=2):
    """Menghitung basis ** eksponen (default: kuadrat)."""
    return basis ** eksponen

print(hitung_pangkat(3))      # 9  (eksponen=2 default)
print(hitung_pangkat(3, 3))   # 27
print(hitung_pangkat(2, 10))  # 1024
```

Nilai default dievaluasi **sekali** saat definisi fungsi, bukan setiap kali dipanggil. Jangan gunakan objek *mutable* (seperti `list`) sebagai nilai default karena bisa menyebabkan bug halus.

**Salah:** `def tambah(data, daftar=[]):` --- daftar yang sama dipakai ulang!\\
**Benar:** `def tambah(data, daftar=None): if daftar is None: daftar = []`

## Keyword Arguments

Saat memanggil fungsi, argumen dapat dilewatkan dengan nama parameter secara eksplisit (Tutorial Python §4.9.2):

```python
def cetak_info(nama, umur, kota="Jakarta"):
    print(f"{nama}, {umur} tahun, dari {kota}")

# Pemanggilan dengan keyword args
cetak_info("Budi", 20)
cetak_info("Sari", umur=22, kota="Bandung")
cetak_info(kota="Surabaya", nama="Andi", umur=19)
```

## Argumen Tak Tentu: *args

Untuk fungsi yang menerima jumlah argumen tak tentu (Tutorial Python §4.9.4):

```python
def jumlahkan(*angka):
    """Menjumlahkan semua argumen yang diberikan."""
    total = 0
    for n in angka:
        total += n
    return total

print(jumlahkan(1, 2, 3))           # 6
print(jumlahkan(10, 20, 30, 40))    # 100

def cetak_profil(**info):
    """Mencetak pasangan kunci-nilai."""
    for kunci, nilai in info.items():
        print(f"  {kunci}: {nilai}")

cetak_profil(nama="Rini", nim="12345", ipk=3.8)
```

## Ekspresi Lambda

Lambda adalah fungsi kecil anonim (tanpa nama) yang ditulis dalam satu baris (Tutorial Python §4.9.6):

```python
# Fungsi biasa
def kali_dua(x):
    return x * 2

# Ekuivalen lambda
kali_dua = lambda x: x * 2
print(kali_dua(5))   # 10

# Lambda berguna sebagai argumen ke fungsi lain
nilai = [("Andi", 85), ("Budi", 72), ("Citra", 90)]
nilai.sort(key=lambda item: item[1])   # sort berdasarkan nilai
print(nilai)

# sorted dengan lambda
angka = [5, 1, 8, 3, 9, 2]
terurut = sorted(angka, key=lambda x: -x)  # urutan menurun
print(terurut)   # [9, 8, 5, 3, 2, 1]
```

Lambda terbatas pada **satu ekspresi**. Untuk logika yang lebih kompleks, gunakan fungsi biasa dengan `def`. Lambda paling berguna saat dipakai sebagai argumen pada fungsi seperti `sorted()`, `map()`, `filter()`.

## Anotasi Fungsi (Type Hints)

Python mendukung anotasi tipe untuk mendokumentasikan tipe parameter dan nilai balik (Tutorial Python §4.9.8). Anotasi tidak mempengaruhi eksekusi program, tetapi membantu pembaca dan alat analisis statis:

```python
def hitung_luas(panjang: float, lebar: float) -> float:
    """Menghitung luas persegi panjang."""
    return panjang * lebar

def sapa(nama: str, formal: bool = False) -> str:
    """Membuat kalimat salam."""
    if formal:
        return f"Selamat pagi, {nama}."
    return f"Halo, {nama}!"

print(hitung_luas(5.0, 3.0))     # 15.0
print(sapa("Budi", formal=True))  # Selamat pagi, Budi.
```

## Ruang Lingkup Variabel

Variabel yang diberi nilai di dalam fungsi bersifat *lokal* kecuali dinyatakan lain dengan `global` atau `nonlocal`. Penggunaan `global` jarang disarankan karena membuat kode sulit dilacak:

```python
x = 10   # variabel global

def ubah_lokal():
    x = 99   # variabel lokal, berbeda dari global x
    print("Dalam fungsi:", x)

ubah_lokal()
print("Di luar fungsi:", x)   # masih 10
```

## List: Lebih Dalam

```python
angka = [3, 1, 4, 1, 5, 9, 2, 6]
angka.append(7)          # tambah di akhir
angka.insert(2, 99)      # sisipkan di indeks 2
angka.remove(1)          # hapus kemunculan pertama nilai 1
dikeluarkan = angka.pop()    # hapus & kembalikan elemen terakhir
dikeluarkan2 = angka.pop(0)  # hapus & kembalikan di indeks 0
angka.sort()             # urut ascending (in-place)
angka.reverse()          # balik urutan (in-place)
salinan = angka.copy()   # buat salinan dangkal
print(angka.count(1))    # hitung kemunculan nilai 1
print(angka.index(5))    # indeks pertama nilai 5
```

### List sebagai Stack (LIFO)

Stack menggunakan `append()` untuk push dan `pop()` untuk pop (Tutorial Python §5.1.1):

```python
stack = []
stack.append("halaman_1")
stack.append("halaman_2")
stack.append("halaman_3")
print(stack.pop())   # "halaman_3" -- LIFO
print(stack.pop())   # "halaman_2"
```

### List sebagai Queue (FIFO)

Untuk queue efisien, gunakan `collections.deque` karena `list.pop(0)` lambat untuk list besar (Tutorial Python §5.1.2):

```python
from collections import deque
antrian = deque(["pelanggan_1", "pelanggan_2", "pelanggan_3"])
antrian.append("pelanggan_4")     # masuk di belakang
print(antrian.popleft())          # keluar dari depan: "pelanggan_1"
print(antrian.popleft())          # "pelanggan_2"
```

### List Comprehension

List comprehension menyediakan cara ringkas untuk membuat list (Tutorial Python §5.1.3):

```python
# Cara lama: loop biasa
kuadrat = []
for i in range(10):
    kuadrat.append(i ** 2)

# List comprehension -- lebih ringkas
kuadrat = [i ** 2 for i in range(10)]
print(kuadrat)   # [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# Dengan kondisi filter
genap_kuadrat = [i ** 2 for i in range(10) if i % 2 == 0]
print(genap_kuadrat)   # [0, 4, 16, 36, 64]

# Manipulasi string
kata = ["  apel  ", "  JERUK  ", " mangga "]
bersih = [k.strip().lower() for k in kata]
print(bersih)   # ['apel', 'jeruk', 'mangga']
```

## Set

Koleksi unik tanpa urutan tetap; berguna untuk menghapus duplikasi atau uji keanggotaan cepat (Tutorial Python §5.4):

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A | B)   # union:        {1,2,3,4,5,6,7,8}
print(A & B)   # intersection: {4, 5}
print(A - B)   # difference:   {1, 2, 3}
print(A ^ B)   # symmetric diff: {1,2,3,6,7,8}

# Cek keanggotaan O(1)
print(3 in A)     # True
print(9 not in A) # True

# Hapus duplikasi dari list
data = [1, 2, 2, 3, 3, 3, 4]
unik = list(set(data))
print(sorted(unik))   # [1, 2, 3, 4]
```

### Kode: matriks_3x3.py

```python
"""Matriks 3x3: isi dan cetak."""

# mat[i][j] untuk i,j in 1..3 - baris/kolom 0 tidak dipakai.


def isi(matrik: list[list[int]]) -> None:
    for i in range(1, 4):
        for j in range(1, 4):
            matrik[i][j] = int(input(f"A[{i},{j}] = "))


def cetak(matrik: list[list[int]]) -> None:
    for i in range(1, 4):
        for j in range(1, 4):
            print(f"{matrik[i][j]:3d}", end="")
        print()


def main() -> None:
    matrik = [[0] * 4 for _ in range(4)]
    print("Mengisi elemen matrik A")
    isi(matrik)
    print("Mencetak elemen matrik A")
    cetak(matrik)


if __name__ == "__main__":
    main()
```

### Kode: daf_pegawai_array.py

```python
"""Daftar 5 pegawai: input lalu cetak satu per satu."""

from dataclasses import dataclass


@dataclass
class Pegawai:
    nama: str = ""
    gol: str = " "
    jml_anak: int = 0


def input_pegawai(pegawai: Pegawai) -> None:
    pegawai.nama = input("Nama        : ")
    pegawai.gol = input("Gol (A - D) : ").strip().upper()[:1]
    pegawai.jml_anak = int(input("Jumlah Anak : "))


def upah_kotor(gol: str) -> float:
    g = gol.upper()
    if g == "A":
        return 1_000_000.0
    if g == "B":
        return 800_000.0
    if g == "C":
        return 600_000.0
    if g == "D":
        return 400_000.0
    return 0.0


def persen_tunjangan(jml_anak: int) -> float:
    if jml_anak > 2:
        return 0.3
    return 0.2


def cetak(pegawai: Pegawai) -> None:
    print("Nama        :", pegawai.nama)
    print("Gol (A - D) :", pegawai.gol)
    print("Jumlah Anak :", pegawai.jml_anak)
    uk = upah_kotor(pegawai.gol)
    upah_bersih = uk - (uk * persen_tunjangan(pegawai.jml_anak))
    print(f"Upah        : {upah_bersih:5.2f}")


def main() -> None:
    daf_pegawai = [Pegawai() for _ in range(5)]
    for i in range(5):
        print("Data ke ", i + 1, ":", sep="")
        input_pegawai(daf_pegawai[i])
    for i in range(5):
        print("Data ke ", i + 1, ":", sep="")
        cetak(daf_pegawai[i])
        input("Tekan Enter untuk melanjutkan...")


if __name__ == "__main__":
    main()
```

## Metode Objek Berkas Lanjut

Objek berkas memiliki beberapa metode yang berguna (Tutorial Python §7.2.1):

```python
with open("data.txt", "w", encoding="utf-8") as f:
    for i in range(1, 6):
        f.write(f"Baris {i}: data penting\n")

with open("data.txt", "r", encoding="utf-8") as f:
    baris1 = f.readline()          # baca satu baris
    print(repr(baris1))            # 'Baris 1: data penting\n'

    semua = f.readlines()          # baca semua sisa sebagai list
    print(len(semua))              # 4 (baris 2--5)

# seek dan tell: navigasi posisi berkas
with open("data.txt", "rb") as f:    # buka mode binary untuk seek
    print(f.tell())      # 0 -- posisi awal
    f.seek(10)           # pindah ke byte ke-10
    print(f.tell())      # 10
    f.seek(0)            # kembali ke awal
    print(f.read(5))     # baca 5 byte pertama
```

## Penanganan Kesalahan Lanjut: try

Python menggunakan *exception* untuk menangani kesalahan runtime (Tutorial Python §8):

```python
def baca_angka(prompt: str) -> int:
    """Membaca angka bulat dari pengguna dengan validasi."""
    while True:
        try:
            nilai = int(input(prompt))
        except ValueError:
            print("Masukan tidak valid. Masukkan angka bulat.")
        else:
            # Dieksekusi jika try BERHASIL (tidak ada exception)
            return nilai

# Beberapa jenis exception sekaligus
def buka_berkas(nama: str) -> str:
    try:
        with open(nama, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Berkas '{nama}' tidak ditemukan.")
        return ""
    except PermissionError:
        print(f"Tidak ada izin membaca '{nama}'.")
        return ""
    except OSError as e:
        print(f"Kesalahan OS: {e}")
        return ""
    finally:
        # SELALU dieksekusi, terlepas dari hasil
        print(f"Percobaan membuka '{nama}' selesai.")
```

```python
def bagi(a: float, b: float) -> float:
    if b == 0:
        raise ValueError("Pembagi tidak boleh nol!")
    return a / b

try:
    hasil = bagi(10, 0)
except ValueError as e:
    print(f"Error: {e}")   # Error: Pembagi tidak boleh nol!
```

**Alur `try**/`except`/`else`/`finally`:`
- **`try**:` Kode yang mungkin memunculkan exception
- **`except**:` Menangani exception jika terjadi
- **`else**:` Dieksekusi *hanya jika* try berhasil tanpa exception
- **`finally**:` *Selalu* dieksekusi (untuk cleanup, menutup koneksi, dll.)

## Studi Kasus dan Contoh Kode

Berikut adalah contoh-contoh program Python yang mengimplementasikan konsep-konsep pada modul ini.

### Kode: baca_tulis.py

```python
"""Menulis dan membaca file teks dengan penanganan kesalahan dasar."""


def main() -> None:
    nama_file = "data_mahasiswa.txt"
    
    # Menulis ke file
    try:
        with open(nama_file, "w", encoding="utf-8") as f:
            f.write("Budi, 3.8\n")
            f.write("Siti, 3.9\n")
        print(f"Data berhasil ditulis ke {nama_file}.")
    except IOError as e:
        print(f"Gagal menulis file: {e}")

    # Membaca dari file
    try:
        with open(nama_file, "r", encoding="utf-8") as f:
            print("\nIsi file:")
            for baris in f:
                print(baris.strip())
    except FileNotFoundError:
        print(f"File {nama_file} tidak ditemukan.")

if __name__ == "__main__":
    main()
```

### Kode: olah_json.py

```python
"""Menyimpan dan memuat data JSON."""
import json


def main() -> None:
    data = {
        "mata_kuliah": "Praktikum Pemrograman",
        "peserta": ["Andi", "Budi", "Citra"]
    }
    file_json = "data.json"
    
    with open(file_json, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)
        print("Data JSON telah disimpan.")
        
    with open(file_json, "r", encoding="utf-8") as f:
        data_baca = json.load(f)
        print("Data JSON dibaca:", data_baca["mata_kuliah"])

if __name__ == "__main__":
    main()
```

```python
# Debug dengan print (sementara)
def cari_maks(daftar):
    print(f"DEBUG: input={daftar}")   # hapus setelah debug
    if not daftar:
        return None
    maks = daftar[0]
    for x in daftar[1:]:
        if x > maks:
            maks = x
    print(f"DEBUG: maks={maks}")      # hapus setelah debug
    return maks
```

## Konsep Kelas dan Objek

**Kelas** adalah cetak biru; **objek** adalah instansi. Di Python, atribut biasanya diinisialisasi di `__init__`.

```python
class Mahasiswa:
    def __init__(self, nim, nama):
        self.nim = nim
        self.nama = nama

    def perkenalan(self):
        return f"{self.nim} - {self.nama}"

m = Mahasiswa("12345", "Andi")
print(m.perkenalan())
```

## Pewarisan Sederhana

Anak mewarisi metode induk dan dapat menimpa (*override*).

```python
class Pengguna:
    def __init__(self, username):
        self.username = username

class Admin(Pengguna):
    def __init__(self, username, level):
        super().__init__(username)
        self.level = level

    def info(self):
        return f"{self.username} (admin level {self.level})"
```

## Studi Kasus dan Contoh Kode

Berikut adalah contoh-contoh program Python yang mengimplementasikan konsep-konsep pada modul ini.

### Kode: kendaraan.py

```python
"""Contoh kelas dasar dan pewarisan."""


class Kendaraan:
    def __init__(self, merk: str, tahun: int) -> None:
        self.merk = merk
        self.tahun = tahun

    def info(self) -> str:
        return f"{self.merk} ({self.tahun})"

class Mobil(Kendaraan):
    def __init__(self, merk: str, tahun: int, jumlah_pintu: int) -> None:
        super().__init__(merk, tahun)
        self.jumlah_pintu = jumlah_pintu

    def info(self) -> str:
        dasar = super().info()
        return f"{dasar} - {self.jumlah_pintu} pintu"

def main() -> None:
    v1 = Kendaraan("Yamaha", 2015)
    v2 = Mobil("Toyota", 2022, 4)
    print(v1.info())
    print(v2.info())

if __name__ == "__main__":
    main()
```

# Modul 10: Comprehension, Generator, dan Iterator Pengenalan

- Sub-CPMK 7.2: Menulis list/dict/set comprehension dan fungsi generator sederhana dengan `yield`
 untuk menghasilkan deret data secara hemat memori

## Sarana dan prasarana

Python 3.10+, editor, terminal; tidak memerlukan paket pihak ketiga.

## Keselamatan kerja dan etika

Etika akademik: comprehension dan generator harus Anda pahami, bukan hanya disalin tanpa memahami alur data.

## Prosedur kerja

- Pelajari sintaks comprehension dari contoh.
- Bandingkan dengan loop klasik untuk kasus yang sama.
- Tulis satu generator dengan `yield`; iterasikan dengan `for` atau `next()`.
- Refaktor potongan kode loop panjang menjadi gaya idiomatik (aktivitas).

## List dan Dict Comprehension

```python
squares = [x * x for x in range(10) if x % 2 == 0]
nama_ke_id = {n.lower(): i for i, n in enumerate(["Ali", "Budi", "Ali"])}
```

## Generator dengan yield

```python
def countdown(n):
    while n >= 0:
        yield n
        n -= 1

for x in countdown(3):
    print(x)
```

## Studi Kasus dan Contoh Kode

Berikut adalah contoh-contoh program Python yang mengimplementasikan konsep-konsep pada modul ini.

### Kode: comprehension.py

```python
"""Contoh penggunaan list comprehension dan generator."""


def deret_genap(batas: int):
    """Generator yang menghasilkan bilangan genap hingga batas."""
    for i in range(batas + 1):
        if i % 2 == 0:
            yield i

def main() -> None:
    # List comprehension
    angka = [1, 2, 3, 4, 5]
    kuadrat = [x**2 for x in angka if x % 2 != 0]
    print(f"Kuadrat ganjil: {kuadrat}")
    
    # Generator
    genap = deret_genap(10)
    print("Deret genap 0-10:", list(genap))

if __name__ == "__main__":
    main()
```

# Modul 11: Pengujian Unit dengan unittest

## Sarana dan prasarana

Python 3.10+ dengan pustaka standar `unittest`; `pytest` opsional (`pip install pytest` di `venv` laboratorium jika diizinkan).

## Keselamatan kerja dan etika

## Tes dengan unittest

```python
import unittest

def tambah(a, b):
    return a + b

class TestTambah(unittest.TestCase):
    def test_positif(self):
        self.assertEqual(tambah(2, 3), 5)

    def test_negatif(self):
        self.assertEqual(tambah(-1, 1), 0)

if __name__ == "__main__":
    unittest.main()
```

## Studi Kasus dan Contoh Kode

Berikut adalah contoh-contoh program Python yang mengimplementasikan konsep-konsep pada modul ini.

### Kode: test_matematika.py

```python
"""Contoh pengujian unit sederhana dengan unittest."""
import unittest


def tambah(a: int, b: int) -> int:
    return a + b

def bagi(a: float, b: float) -> float:
    if b == 0:
        raise ValueError("Pembagi tidak boleh nol")
    return a / b

class TestMatematika(unittest.TestCase):
    def test_tambah(self) -> None:
        self.assertEqual(tambah(2, 3), 5)
        self.assertEqual(tambah(-1, 1), 0)

    def test_bagi(self) -> None:
        self.assertEqual(bagi(10, 2), 5.0)
        with self.assertRaises(ValueError):
            bagi(10, 0)

if __name__ == "__main__":
    unittest.main()
```
